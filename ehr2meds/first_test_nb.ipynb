### First nb to get an overview of project specific registries

## Registers

Skin care specific
- nmsc (The Danish Nonmelanoma Skin Cancer Dermatology Database)
- pato (Patobank)
- mohs (Mohs micrographic survery database)

DST general 
- bef (Befolkning)
- vnds (Historiske vandringer)

## Selected Variables for each register

## nmsc
- 'CPR' (Person ID, is noted wrong, is actually PNR)
    - Format:   NNNNNNNNNNNN
    - dtype:    Char  
    - length:   12
- 'BehandlingsDato' (Date of treatment, interchagable with diagnosis day)
    - Format:   NNLLLNNNN (DDMMMYYYY)
    - dtype:    DATE9.
    - length:   8 (9?)
- 'KliniskVurderingTumorType' (Type of tumor, determined by clinician) 
    - Format:   L*_L*
    - dtype:    Char
    - length:   Up to 40 
- 'HistologiskVurderingTumorType' (Type of tumor, determined through histology)
    - Format:   L*_L*
    - dtype:    Char
    - length:   Up to 40 
- 'GruppeId' (Tumor ID)
    - Format:   Mix of D and N
    - dtype:    Char
    - length:   72

## pato 
- 'pnr' (Person ID)
    - Format:   NNNNNNNNNNNN
    - dtype:    Char
    - length:   12  
- 'C_SNOMEDKODE' (Identifier for a specific clinical concept)
    - Format:   LNNNNN 
    - dtype :   Char
    - length:   6 
- 'D_REKVDATO' (The date the examination was ordered)
    - Format:   NNNNNNNN (DDMMYYYY)   
    - dtype:    Char 
    - length:   8 

## mohs
- 'pnr' (Person ID)
    - Format:   NNNNNNNNNNNN
    - dtype:    Char
    - length:   12  
- 'biopsy_date'
    - Format:   DDDD-DD-DD (YYYY-MM-DD)
    - dtype:    Char
    - length:   10
- 'sex'
    - Format:   D
    - dtype:    Char
    - length:   1
- 'skin_type' 
    - Format:  D*
    - dtype:   Char 
    - length:  up to 3
